# Getting Started with Python for Data Analysis

Python programming language has numerous functions. These functions are organized into modules. **Modules** are python scripts that contain Python objects and functions. In order to use the objects and functions inside these modules (or commonly libraries), we must first **import** them. 

`Pandas` (Python for Data Analysis) is a popular python library. We will use this library extensively in this workshop. This library is automatically installed during Anaconda installation. It is also installed and available on Google Colab or UofT Jupyter Hub. 

Other popular python libraries are `NumPy` and `matplotlib` used for scientific computing and visualization, respectively. We will import these libraries too. 

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

Let's take a moment to understand the three lines of code above. 

First `import` itself is a function which is available directly in Python. This function is used to import or bring the functions of a Python module into the Python environment we are working in. The import function is followed by the module we are looking to import. So, here we have imported `pandas`, `numpy` and `matplotlib` libraries. 

Often instead of importing the libraries with its original name, we import them with an alias using `as` function. This is useful because everytime we want to use a function inside the library, we can type just the short alias instead of typing the full name of the module. So `pd` for `pandas` and `np` for numpy and so on. 

Notice that the last line of import is slightly different in that it contains a `.`, which is the **dot operator**. First, dot operators are used to access items that are inside of the object following it. Second, sometimes the libraries are so big that they need to managed inside multiple modules. Each of the sub-modules of a module can be accessed by using the dot operator. In this example, `pyplot` sub-module is imported instead of the entire `matplotlib` module. Then the alias `plt` is given to this sub-module. Throughout this tutorial, we will make use of the dot operator to access the functions available to an object.

It seems that some routes have multiple route names. To find out which routename have the same route, we can first isolate the two columns and then apply `drop_duplicates` method to remove any duplicated rows. In the resulting dataframe, we can apply `duplicated` method, which will result in Boolean (True or False) `Series`. 

A `Series` is a Pandas data structure designed to store information of one variable only. If there are more than one variable it's a `DataFrame`.

```python
allroutes = data[['route','routename']].drop_duplicates()
allroutes
```

```python
allroutes.duplicated(['route'])
```

This resulting series can now be used to filter the rows of our interest. If the series is False, the row is not selected and if it is True the row will be selected. 

Earlier we learnt how to select columns and now this is about selecting rows. In both cases we need to use the index operator,`[ ]`. Inside the index operator we can insert this Boolean series. The `duplicated` method offers a `keep` parameter. We will set this parameter's value to False so that we can see both routenames. 

```python
duplicated = allroutes[allroutes.duplicated(['route'], keep=False)]
duplicated
```
